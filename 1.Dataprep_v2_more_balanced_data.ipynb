Import important libraries

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import joblib
sns.set()
%matplotlib inline
```

Load data then show .head & describe them.

```python
df = pd.read_csv('fifa20.csv')
```

## Checking NaN value

```python
df_clean = df.copy()
nan_percentage = df_clean.isnull().sum().sort_values(ascending=False) / df_clean.shape[0]
missing_val = nan_percentage[nan_percentage > 0]

plt.figure(figsize=(16,6))
sns.barplot(x=missing_val.index.values, y=missing_val.values * 100, palette="Reds_r");
plt.title("Percentage of missing values in data");
plt.ylabel("%");
plt.xticks(rotation=90);
```

```python
unwanted_feature = [
    'loaned_from', 'nation_position', 'nation_jersey_number',
    'player_tags', 'gk_positioning', 'gk_diving', 'gk_handling',
    'gk_kicking', 'gk_reflexes', 'gk_speed', 'player_traits',
    'joined', 'contract_valid_until', 'team_jersey_number', #also drop not important feature
    'player_url', 'short_name', 'long_name','dob', 'nationality', 'club',
    'real_face', 'ls', 'st', 'rs',
    'lw', 'lf', 'cf', 'rf', 'rw', 'lam', 'cam', 'ram', 'lm', 'lcm', 'cm',
    'rcm', 'rm', 'lwb', 'ldm', 'cdm', 'rdm', 'rwb', 'lb', 'lcb', 'cb',
    'rcb', 'rb',
]
df_clean = df_clean.drop(unwanted_feature, axis=1)
```

```python
missing_val.index.values
```

Fill all missing data with 0 except for team_position i will fill using special char "-"

```python
df_clean['team_position'].fillna('-', inplace=True)
```

```python
df_clean = df_clean.fillna(0) #replace all missing data with 0
```

### Binning data

```python
len(df_clean["player_positions"].unique())
```

```python
df_clean['player_positions'][0] #lionel messi have 3 player_positions
```

player_positions have a lots of classes and some player have combination of many position.

I assume the first position is the best position.

```python
df_clean['best_pos'] = df_clean['player_positions'].str.split(',').str[0]
df_player = df_clean.copy() #make new dataframe
```

Plot the best_pos distribution

```python
plt.figure(figsize=(16,6))
# mpl.rcParams['figure.dpi'] = 100
f = sns.countplot(x='best_pos', data=df_player)
f.set_title("best_pos distribution")
plt.show()
```

The data looks cleaner

Map position to *8 classes/Category*, based on this https://en.wikipedia.org/wiki/Association_football_positions#Centre_forward

```python
df_player['best_pos'] = df_player['best_pos'].map({
    'ST': 'Fwd', #Striker
    'CF': 'Fwd', #Center Foward
    'LW': 'Fwd', #left wing
    'RW': 'Fwd', #Right wing
    'CM': 'Mid_Centre', #Centre midfield
    'CAM': 'Mid_Centre', #Centre ATTACK midfield
    'CDM': 'Mid_Centre', #Centre Deffensive midfield
    'LM': 'Mid_Side',  #Left mid side
    'RM': 'Mid_Side', #Right mid side
    'CB': 'Back', #Center Back
    'LB': 'Back', #Left Back
    'RB': 'Back', #Right Back
    'LWB': 'Back', #Left wing back
    'RWB': 'Back', #Right Wing back
    'GK': 'Goal_Keeper',}) #Goalkeeper ofc
classes = ['Back',
    'Fwd',
    'Mid_Centre',
    'Mid_Side',
    'Goal_Keeper']
df_player = df_player.drop(['player_positions'], axis=1) #drop player positions, we don't need it anymore
```

```python
plt.figure(figsize=(16,6))
# mpl.rcParams['figure.dpi'] = 100
f = sns.countplot(x='best_pos', data=df_player)
f.set_title("best_pos distribution")
plt.show()
```

Because our data contains non-numeric data, encode using LabelEncoder()

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
must_labeled = ['preferred_foot', 'work_rate', 'body_type' , 'team_position', 'best_pos']
df_player_labeled = df_player.copy()
```

```python
from collections import defaultdict
d = defaultdict(LabelEncoder)
# Encoding the variable
fit = df_player_labeled[must_labeled].apply(lambda x: d[x.name].fit_transform(x))
```

```python
df_player_labeled[must_labeled] = df_player_labeled[must_labeled].apply(lambda x: d[x.name].transform(x))
```

```python
df_player_labeled.to_csv('data/main_dataset_labeled_v2.csv')
joblib.dump(d,'data/label_encoder_model_v2.sav') #save labelEncoder instance for future use
```

## Handling outliers

remove outlier using Z-Score.

https://towardsdatascience.com/ways-to-detect-and-remove-the-outliers-404d16608dba

```python
from scipy import stats
z = np.abs(stats.zscore(df_player_labeled))
print(z) #print all z-scores
```

```python
threshold = 4
df_player_labeled_no_outliers = df_player_labeled[(z < threshold).all(axis=1)]
```

```python
print('Total data without outliers = ', len(df_player_labeled_no_outliers))
print('Total data without Handling outliers = ',len(df_player_labeled))
```

```python
df_player_labeled_no_outliers.to_csv('data/main_dataset_no_outliers_v2.csv')
```

```python
plt.figure(figsize=(16,6))
# mpl.rcParams['figure.dpi'] = 100
f = sns.countplot(x='best_pos', data=df_player_labeled_no_outliers)
f.set_title("best_pos distribution")
plt.show()
```
